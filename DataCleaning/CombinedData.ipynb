```python
import pandas as pd
```

```python
#Loading data

# Read the first CSV file into a DataFrame
df1 = pd.read_csv('Cleaned_Apartments.csv')

# Read the second CSV file into another DataFrame
df2 = pd.read_csv('Cleaned_BuyRent.csv')
```

```python
df1.head()
```

```python
df2.head()
```

```python
#Re-arranging the columns in the df1
# Rearrange the order of columns
new_order = ['City', 'Location', 'Bedrooms', 'Bathrooms', 'Price']  # Specify the desired column order

# Create a new DataFrame with columns in the desired order
df3 = df1[new_order]
```

```python
df3.head()
```

```python
# Concatenate the two DataFrames along rows (axis=0)
combined_df = pd.concat([df2, df3], axis=0)
```

```python
combined_df.head()
```

### Creating a csv file of the clean BuyRent Data

```python
# Save the combined DataFrame to a new CSV file
combined_df.to_csv('Combined_df.csv', index=False)
```
